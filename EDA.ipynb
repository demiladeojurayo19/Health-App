This command furnishes the mean, standard deviation, and a five-number summary (Minimum, Lower Quartile (Q1) = 25%, Median (Q2) = 50%, Upper Quartile (Q3) = 75%, Maximum) for each numerical feature. Additionally, this summary is instrumental in constructing a box plot.

## Additional statistical measures that will be calculated are kurtosis and skew.

Both kurtosis and skew are important statistical measures . Kurtosis is the measure of outliers present in the data. High kurtosis (>3) indicates a large number of outliers and low kurtosis (<3) a lack of outliers. Skew will indicate how symmetrical our data is.

```python
eda_module.display_skewness(data)
```

```python
eda_module.display_kurtosis(data)
```
